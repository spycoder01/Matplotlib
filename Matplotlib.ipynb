```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
```

```python
#ploting a 2D plot
price = [23000,23500,26000,25500,24000]
year = [2021,2022,2023,2024,2025]
plt.plot(year,price)
```

```python
#plotting curve
Batsman = pd.read_csv('/content/sharma-kohli.csv')
Batsman
plt.plot(Batsman['index'],Batsman['RG Sharma'])
plt.plot(Batsman['index'],Batsman['V Kohli'])
plt.title('Rohit vs Virat comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#colour of curves
plt.plot(Batsman['index'],Batsman['RG Sharma'],color='red',linestyle='solid',linewidth='2')
plt.plot(Batsman['index'],Batsman['V Kohli'],color='blue',linestyle='dashed',linewidth='1')
plt.title('Rohit vs Virat comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

#use can choose any colour code using hex code
```

```python
#show points
plt.plot(Batsman['index'],Batsman['RG Sharma'],color='red',linestyle='solid',marker='D',markersize=3)
plt.plot(Batsman['index'],Batsman['V Kohli'],color='blue',linestyle='dashed',marker='o',markersize=4)
plt.title('Rohit vs Virat comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#curve labeling
plt.plot(Batsman['index'],Batsman['RG Sharma'],color='#0f2d61',linestyle='dashed',linewidth=2,label='Rohit',marker='d')
plt.plot(Batsman['index'],Batsman['V Kohli'],color='red',label='Virat',marker='o',markersize=4)
plt.title('Rohit vs Virat comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid()
```

```python
#trim graph_limiting of axes
price = [23000,23500,26000,25500,24000,80000]
year = [2021,2022,2023,2024,2025,2026]
plt.plot(year,price,color='red',marker='o',linewidth='1')
plt.title('Rohit vs Virat comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.xlim(2021,2026)
plt.ylim(20000,30000)
plt.grid()
plt.show()
```

Scatter_Plots

```python
#plot.scatter simple function
x = np.linspace(-10,10,50)
y = 10*x + np.random.randint(0,300,50)
y
plt.scatter(x,y,s=25)
```

```python
df = pd.read_csv('/content/batter.csv')
df = df.head(50)
df
plt.scatter(df['avg'],df['strike_rate'],color='red')
plt.title('Avg and Strike rate of top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
# size
tips = sns.load_dataset('tips')
print(tips.head(10))

# slower
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*10)
plt.title('Tips and total bill of 10 customers')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

Bar Chart

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='#085264')
```

```python
# horizontal bar chart
plt.barh(colors,children,color='#085264')
```

```python
df = pd.read_csv('/content/batsman_season_record.csv')
df
plt.barh(df['batsman'],df['2015'],color='#7f3400')
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']),label='2015')
plt.legend()
plt.show()
```

```python
# Histogram
plt.hist(df['avg'],bins=[20,25,30,35,40,45,50])
plt.show()
```

```python
# Pie chart
data = [71,86,67,84]
subjects = ['phy','Eng','chem','maths']
plt.pie(data,labels=subjects,autopct='%0.1f%%',explode=[0.1,0,0,0],shadow=True)
# plt.show()
```
